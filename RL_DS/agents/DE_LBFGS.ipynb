```python

#Differential Evolution
from scipy.optimize import differential_evolution
time=47
bounds = [(0, 500)] * time


def objective_de(actions):
  env=RetailerOrders()
  env.run(actions)
  return env.retailer_total_cost

result_de = differential_evolution(objective_de, bounds, seed=0, disp=True,maxiter=1000)

print("Best parameters found by Differential Evolution:")
print(result_de.x)
print("Best cost found by Differential Evolution:")
print(-result_de.fun)

```

```python
env=RetailerOrders()
env.run(result_de.x)
pd.DataFrame(env.history)
```
